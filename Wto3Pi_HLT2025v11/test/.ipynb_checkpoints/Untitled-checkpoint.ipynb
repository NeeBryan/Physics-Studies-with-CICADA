```python
import uproot
import awkward as ak
import numpy as np
import matplotlib.pyplot as plt
import mplhep as hep
hep.style.use("CMS")

# Load ROOT file
file = uproot.open("CICADA_HIG-RunIISummer20UL17HLT_2025v11-13144.root")
tree = file["Events"]

# Define triggers of interest
trigger_branches = {
    "DST_PFScouting_CICADAVLoose": "DST_PFScouting_CICADAVLoose",
    "L1_SingleJet46er2p5": "L1_SingleJet46er2p5_NotBptxOR_3BX",
    "L1_SingleMuOpen_er1p4": "L1_SingleMuOpen_er1p4_NotBptxOR_3BX",
    "L1_SingleMuOpen_er1p1": "L1_SingleMuOpen_er1p1_NotBptxOR_3BX",
    "L1_SingleJet43er2p5": "L1_SingleJet43er2p5_NotBptxOR_3BX",
    "L1_SingleEG60": "L1_SingleEG60",
}

# Load branches
branches_to_load = list(trigger_branches.values()) + [
    "Jet_pt", "Jet_eta", "Jet_phi", "Jet_mass", "nJet"
]
arrays = tree.arrays(branches_to_load)

# Function to plot Jet variables per trigger
def plot_jet_variable(var_name, xlabel, filename):
    plt.figure(figsize=(8,6))
    for label, trig_branch in trigger_branches.items():
        mask = arrays[trig_branch]
        jets = arrays[var_name][mask]
        jets_flat = ak.flatten(jets)
        if len(jets_flat) == 0:
            print(f"WARNING: No jets found for trigger {label}")
            continue
        plt.hist(jets_flat, bins=50, histtype='step', label=label, density=True)
    
    plt.xlabel(xlabel, fontsize=20)
    plt.ylabel("a.u. (Normalized)", fontsize=20)
    plt.title(f"{xlabel} Distribution by Trigger", fontsize=25)
    plt.legend(fontsize=15)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    plt.savefig(filename)
    plt.close()

# Plot Jet variable comparisons
plot_jet_variable("Jet_pt", "Jet $p_T$ [GeV]", "jet_pt_by_trigger.png")
plot_jet_variable("Jet_eta", "Jet $\eta$", "jet_eta_by_trigger.png")
plot_jet_variable("Jet_phi", "Jet $\phi$", "jet_phi_by_trigger.png")
plot_jet_variable("Jet_mass", "Jet Mass [GeV]", "jet_mass_by_trigger.png")
```
